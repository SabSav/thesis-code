```python
import sys, os
sys.path.append(f'../../code')
path_to_json = os.getcwd()

from IPython.display import display, Markdown
```

```python
import json
def load_case(path):
    display(Markdown(f'## {path}'))
    with open(path, 'r+') as file: bundle = json.load(file)
    print('Temperature:', bundle['temperature'])
    return bundle

display(Markdown('# Files'))
json_files_mc = sorted([pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json') and pos_json.startswith('mc-lT')],
                       reverse = True)
json_files_alg1 = sorted([pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json') and pos_json.startswith('a1-lT')],
                         reverse = True)
json_files_alg2 = sorted([pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json') and pos_json.startswith('a2-lT')],
                         reverse = True)

print("MC: ", json_files_mc)
print("A1: ", json_files_alg1)
print("A2: ", json_files_alg2)

mc = load_case(json_files_mc[0])
alg1 = load_case(json_files_alg1[0])
alg2 = load_case(json_files_alg2[0])
```

```python
import ising
from ising import *
import matplotlib.pyplot as plt
import seaborn as sns

fig_kde, axes = plt.subplots(1, 2)
h1 = sns.kdeplot(data=mc["energy_sample"], bw_method='scott', color='m', ax=axes[0])
h1.set(xlabel=None)
h2 = sns.kdeplot(data=alg1["energy_sample"], bw_method='scott', color='b', ax=axes[0])
h2.set(xlabel=None)
h3 = sns.kdeplot(data=alg2["energy_sample"], bw_method='scott', color='y', ax=axes[0])
h3.set(xlabel=None)
m1 = sns.kdeplot(data=mc["magnetization_sample"], bw_method='scott', color='m', ax=axes[1])
m1.set(xlabel=None)
m2 = sns.kdeplot(data=alg1["magnetization_sample"], bw_method='scott', color='b', ax=axes[1])
m2.set(xlabel=None)
m3 = sns.kdeplot(data=alg2["magnetization_sample"], bw_method='scott', color='y', ax=axes[1])
m3.set(xlabel=None)

axes[0].set_title("Energy")
axes[0].legend(['MonteCarlo', 'Alg1', 'Alg2'], loc="upper left")
axes[1].set_title("Magnetization")
axes[1].legend(['MonteCarlo', 'Alg1', 'Alg2'], loc="upper right")
```

```python
chain = Chain(size=alg1["number of spins"])
bin_edges_engy_mc, metropolis_engy = hist(chain, mc["energy_sample"], engy_flag=True)
bin_edges_engy_alg1, alg1_engy = hist(chain, alg1["energy_sample"], engy_flag=True)
bin_edges_engy_alg2, alg2_engy = hist(chain, alg2["energy_sample"], engy_flag=True)
fig_engy, axes = plt.subplots(1, 3)
axes[0].hist(metropolis_engy, bins=bin_edges_engy_mc, label='Metropolis', color='m')
axes[1].hist(alg1_engy, bins=bin_edges_engy_alg1, label='Alg1', color='b')
axes[2].hist(alg2_engy, bins=bin_edges_engy_alg2, label='Alg2', color='y')
axes[0].set_title("Energy Montecarlo")
axes[1].set_title("Energy Alg1")
axes[2].set_title("Energy Alg2")
plt.show()
```

```python
bin_edges_m_mc, metropolis_m = hist(chain, mc["magnetization_sample"], engy_flag=False)
bin_edges_m_alg1, alg1_m = hist(chain, alg1["magnetization_sample"], engy_flag=False)
bin_edges_m_alg2, alg2_m = hist(chain, alg2["magnetization_sample"], engy_flag=False)
fig_m, axes = plt.subplots(1, 3)
axes[0].hist(metropolis_m, bins=bin_edges_m_mc, label='Metropolis', color='m')
axes[1].hist(alg1_m, bins=bin_edges_m_alg1, label='Alg1', color='b')
axes[2].hist(alg2_m, bins=bin_edges_m_alg2, label='Alg2', color='y')
axes[0].set_title("Magnetization Montecarlo")
axes[1].set_title("Magnetization Alg1")
axes[2].set_title("Magnetization Alg2")
plt.show()
```
